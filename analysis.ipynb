Phase 2: Clean Data
- Create 'color' column in both DFs
- Combine both dataframes into 1 new DF
- Rename columns

```python
red_wine['color'] = "red"
white_wine['color'] = "white"
red_wine.head()
```

```python
white_wine.head()
```

```python
df = pd.concat([red_wine, white_wine])
display(df)
```

```python
df.rename(
    columns= {
    "fixed acidity":"fixed_acidity", 
    "volatile acidity":"volatile_acidity", 
    "citric acid":"citric_acid", 
    "residual sugar":"residual_sugar", 
    "free sulfur dioxide": "free_sulfur_dioxide", 
    "total sulfur dioxide":"total_sulfur_dioxide"
    }, inplace=True)

df.head()
```

```python
ph_red = df['color'] == 'red'
ph_white = df['color'] == 'white'

avg_ph_red = ph_red.mean()
avg_ph_white = ph_white.mean()

display(avg_ph_red, avg_ph_white)
```

```python
df.groupby(['quality', 'color'], as_index=False)['pH'].mean()
```

```python
# Q1 answer
fig = df.groupby('color').mean(numeric_only=True).quality.plot(kind='bar', title='Avg Wine Quality Per Color', figsize=(5,5))
fig.bar_label(fig.containers[0], label_type='edge')
```

```python
df['pH'].describe()
```

```python
bin_edges = [2.72, 3.11, 3.21, 3.32, 4.01]
bin_names = ['high', 'mod_high', 'medium', 'low']

df['acidity_level'] = pd.cut(df['pH'], bin_edges, labels=bin_names)
df.head()
```

Created a new column 'acidity_levels' and categorized the acidity level of all wines into low, medium, mod_high & high.

```python
fig = df.groupby('acidity_level').mean(numeric_only=True).quality.plot(kind='bar', title='Avg Quality Per Acidity Level')
fig.bar_label(fig.containers[0]);
plt.xlabel('Acidity Level')
plt.ylabel('Quality')
```

Q3: Do wines with higher alcoholic content receive higher rating?

```python
# categorize alcoholic content into 2 groups, low being below the median, and high being above or equal to the median.
# then find the mean quality for each group.
alcohol_median = df['alcohol'].median()
alcohol_median
```

```python
high_alcohol_mean = df.query('alcohol >= @alcohol_median').quality.mean()
low_alcohol_mean = df.query('alcohol < @alcohol_median').quality.mean()

alcohol_content = [high_alcohol_mean, low_alcohol_mean]

fig = sns.barplot(data=df, x=['high', 'low'], y=alcohol_content)
fig.bar_label(fig.containers[0])
plt.title('Average rating per alcohol content')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality');
```

High alcohol content has a better quality rating on average.

Q4: Do sweeter wines have better quality rating?

```python
# split residual_sugar into 2 categories, high and low, using the median, then get the mean quality for each.
sugar_median = df['residual_sugar'].median()
sugar_median
```

```python
high_sugar_mean = df.query('residual_sugar >= @sugar_median').quality.mean()
low_sugar_mean = df.query('residual_sugar < @sugar_median').quality.mean()

sugar_content = [high_sugar_mean, low_sugar_mean]

fig = sns.barplot(data=df, x=['High Sugar', 'Low Sugar'], y=sugar_content)
fig.bar_label(fig.containers[0])
plt.xlabel('Sugar Content')
plt.ylabel('Quality')
plt.title('Quality rating per sugar content');
```

Almost identical quality ratings for both high & low sugar content.
